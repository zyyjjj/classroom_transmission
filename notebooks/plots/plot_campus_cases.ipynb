```python
import matplotlib.pyplot as plt
import matplotlib.dates as md

import pandas as pd

plt.rcParams["font.family"] = 'serif'
```

## Fall 2021

```python
all_students_features = pd.read_csv('../data/data_fa21/study_population_truncated_include_nonsurveilled_fa21.csv', index_col = 0) 
emplid = '0x0001CEED0A3584312155FD3B695D2EB6'
hd_positives_all = all_students_features.query('employee_id_hash == @emplid')[['day_idx','positives_identified_on_this_day']]
```

```python
hd_positives_all.head()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

hd_positives_all["day_idx"] = pd.to_datetime(hd_positives_all["day_idx"], format = '%Y-%m-%d')

ax.plot(
    hd_positives_all["day_idx"], hd_positives_all["positives_identified_on_this_day"]
)

ax.xaxis.set_major_locator(md.MonthLocator(interval = 1))
# ax.xaxis.set_major_formatter(md.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_formatter(md.DateFormatter('%b %Y'))

ax.set_ylabel("Number of new cases identified each day")
ax.set_title("Fall 2021 student cases on Ithaca campus")

fig.savefig("../plots/campus_cases_fa21.pdf")
```

## Spring 2022

```python
all_students_features = pd.read_csv('../data/data_sp22/study_population_truncated_include_nonsurveilled_sp22.csv', index_col = 0) 

emplid = '0x0001CEED0A3584312155FD3B695D2EB6'
hd_positives_all = all_students_features.query('employee_id_hash == @emplid')[['day_idx','positives_identified_on_this_day']]
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

hd_positives_all["day_idx"] = pd.to_datetime(hd_positives_all["day_idx"], format = '%Y-%m-%d')

ax.plot(
    hd_positives_all["day_idx"], hd_positives_all["positives_identified_on_this_day"]
)

ax.xaxis.set_major_locator(md.WeekdayLocator(byweekday = 0))
ax.xaxis.set_major_formatter(md.DateFormatter('%Y-%m-%d'))
# ax.xaxis.set_major_formatter(md.DateFormatter('%b %Y'))
ax.xaxis.set_minor_locator(md.DayLocator(interval = 1))


ax.set_ylabel("Number of new cases identified each day")
ax.set_title("Spring 2022 student cases on Ithaca campus")

fig.savefig("../plots/campus_cases_sp22.pdf")
```
