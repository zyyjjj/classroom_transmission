```python
import sys
sys.path.append('../..')

import matplotlib.pyplot as plt
import numpy as np

from src.run_regression import run_cox_regression
```

## Fall 2021

```python
cox_fa21, avg_cp = run_cox_regression(semester="fa21",verbose=True)
```

```python
avg_cp
```

```python
hazard = [0] + [cox_fa21.baseline_cumulative_hazard_.iloc[2]["baseline hazard"]] + \
    list(np.diff(cox_fa21.baseline_cumulative_hazard_.reindex(
        list(range(cox_fa21.baseline_cumulative_hazard_.index.min(),cox_fa21.baseline_cumulative_hazard_.index.max()+1)),method="ffill").values.squeeze(1)
    ))
print(np.mean(hazard))

print(len(hazard))

plt.rcParams["font.family"] = 'serif'

fig, ax = plt.subplots(figsize = (8, 5))

ax.plot(np.arange(1,105), hazard)
ax.set_title("Fall 2021 study period baseline hazard")
ax.set_xlabel("Day of study period")
ax.set_ylabel("Baseline hazard value")

# fig.savefig("../plots/baseline_hazard_fa21.pdf")
```
