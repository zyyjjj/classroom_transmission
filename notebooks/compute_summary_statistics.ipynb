```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fa_covariates = pd.read_csv('../data/data_fa21/all_students_features_T_e=7_finalized.csv', index_col = 0)
sp_covariates = pd.read_csv('../data/data_sp22/all_students_features_T_e=7_finalized.csv', index_col = 0)
```

```python
fa_subjects = fa_covariates[
    ["employee_id_hash", "current_gender", "academic_career"]
].drop_duplicates()

sp_subjects = sp_covariates[
    ["employee_id_hash", "current_gender", "academic_career"]
].drop_duplicates()
```

```python
fa_res = []
fa_ind = []

fa_ind.append("Total")
fa_res.append(
    {
        "num_persons": len(fa_subjects),
        "% p": 100,
        "num_person_days": len(fa_covariates),
        "% pd": 100,
        "incidence_rate": np.round(fa_covariates["infected_on_this_day"].sum() / len(fa_covariates) * 1000, 2)
    }
)
```

```python

for feature in ["current_gender", "academic_career"]:

    g1 = fa_subjects.groupby(feature).aggregate("count")
    g2 = fa_covariates.groupby(feature).aggregate({"infected_on_this_day": ["mean", "count"]})

    for s in g1.index:
        fa_ind.append(s)
        fa_res.append(
            {
                "num_persons": g1.loc[s]["employee_id_hash"],
                "% p": np.round(g1.loc[s]["employee_id_hash"] / len(fa_subjects) * 100),
                "num_person_days": g2.loc[s]["infected_on_this_day"]["count"],
                "% pd": np.round(g2.loc[s]["infected_on_this_day"]["count"] / len(fa_covariates) * 100),
                "incidence_rate": np.round(g2.loc[s]["infected_on_this_day"]["mean"] * 1000, 2)
            }
        )
```

```python
pd.DataFrame(fa_res, index=fa_ind)
```
